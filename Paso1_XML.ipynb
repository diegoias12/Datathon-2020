```python
import pandas as pd 
import xml.etree.ElementTree as et
```

```python
xtree = et.parse('dblp_1990_2000_clean_utf8.xml')
```

```python
xroot = xtree.getroot()
```

```python
df_cols = ["author", "id_article"]
publicaciones = ["article", "inproceeding", "inproceedings"]
# publicaciones = ["article", "inproceeding"]
rows = []

for node in xroot: 
    s_name = node.tag
        
    s_key = node.attrib.get('key')
        
    s_author = None
    if node.find("author") is not None:
        s_author = node.findall("author")
        
    if s_name in publicaciones \
    and s_author is not None \
    and s_key is not None:
        for author in s_author:
            rows.append({"author": author.text, "id_article": s_key})

out_df = pd.DataFrame(rows, columns = df_cols)
```
